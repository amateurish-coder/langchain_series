```python
# read pdf
pdf_reader = PdfReader("/kaggle/input/daniel-cv/resume.pdf")
```

```python
page = pdf_reader.pages[0]
```

```python
text = ""

# collect all pages
for page in pdf_reader.pages:
    
    # extract text
    page_text = page.extract_text()
    
    # add to text list
    text += page_text
```

```python
text
```

```python
from langchain.text_splitter import CharacterTextSplitter

# generate chunks
def generate_chunks(text):
    
    text_splitter = CharacterTextSplitter(
    separator = "\n",
    chunk_size = 100,
    chunk_overlap  = 20,
    length_function = len,
    is_separator_regex = False,
    )
    
    chunks = text_splitter.split_text(text=text)
    return chunks
```

```python
chunks = generate_chunks(text)
chunks
```

```python
from langchain.embeddings import HuggingFaceInstructEmbeddings
from langchain.vectorstores import FAISS

# define embeddings model (to convert text to numerical format)
embeddings = HuggingFaceInstructEmbeddings(model_name="hkunlp/instructor-xl")

# define a database to store the text chunks
vectorstore = FAISS.from_texts(texts=chunks, embedding=embeddings)
```

```python
user_question = 'what languages do I speak?'
docs = vectorstore.similarity_search(query=user_question)
```
